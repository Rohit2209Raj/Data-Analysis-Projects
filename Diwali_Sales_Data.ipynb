```python
grp = df.groupby(['Gender', 'Product_Category'])['Orders'].count()

grp = grp.reset_index(name='Order_Count')

grp = grp.sort_values(['Gender', 'Order_Count'], ascending=[True, False])

top5_per_gender = grp.groupby('Gender').head(5)
top5_female=top5_per_gender.head(5)

top5_male=top5_per_gender.tail()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=top5_female,
    x='Product_Category',
    y='Order_Count'
)
```

BASED ON AMOUNT PER CATEGORY


```python
grpa = df.groupby(['Gender', 'Product_Category'])['Amount'].sum()

grpa = grpa.reset_index(name='Total_amount')

grpa = grpa.sort_values(['Gender', 'Total_amount'], ascending=[True, False])

top5_per_gendera = grpa.groupby('Gender').head(5)
top5_femalea=top5_per_gendera.head(5)

top5_malea=top5_per_gendera.tail()
```

# Zone and state Wise data

```python
df.head(10)
```

```python
df_zone_order = df.groupby('Zone')['Orders'].sum().reset_index().sort_values(by='Orders', ascending=False)
plt.figure(figsize=(10,6))

ax = sns.barplot(data=df_zone_order, x='Zone', y='Orders')

# Add numbers on top of bars
for i in ax.containers:
    ax.bar_label(i)

plt.title("Number of Orders per Zone")
plt.tight_layout()
plt.show()
```

```python
# Central zone places the most amount of orders
```

<!-- ZOne wise amount spent
 -->

```python
df_zone_order = df.groupby('Zone')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(10,6))

ax = sns.barplot(data=df_zone_order, x='Zone', y='Amount')

# Add numbers on top of bars
for i in ax.containers:
    ax.bar_label(i)

plt.title("Total_amount per Zone")
plt.tight_layout()
plt.show()
```

```python
# zone wise Gender Distribution
```

```python
df_zone_gender=df.groupby(by=['Zone','Gender']).size().reset_index(name='count')
df_zone_gender
```

```python
sns.barplot(x='Zone',y='count',hue='Gender',data=df_zone_gender)
```

```python
# Zone wise most popular category
```

```python
df_zone_category=df.groupby(by=['Zone','Product_Category']).size().reset_index(name='count')
df_zone_category
idx = df_zone_category.groupby('Zone')['count'].idxmax()

# # # Select only the rows with the max count per Zone
df_best_per_zone = df_zone_category.loc[idx].reset_index(drop=True)

print(df_best_per_zone)
```

<!-- Zone wise amount -->

state wise orders data

```python
df_state_order = df.groupby('State')['Orders'].sum().reset_index().sort_values(by='Orders', ascending=False)
plt.figure(figsize=(25,6))

ax = sns.barplot(data=df_state_order, x='State', y='Orders')

# Add numbers on top of bars
for i in ax.containers:
    ax.bar_label(i)

plt.title("Number of Orders per State")
plt.tight_layout()
plt.show()
```

<!-- states vs amount -->

```python
df_state_order = df.groupby('State')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(25,6))

ax = sns.barplot(data=df_state_order, x='State', y='Amount')

# Add numbers on top of bars
for i in ax.containers:
    ax.bar_label(i)

plt.title("Total_amount per State")
plt.tight_layout()
plt.show()
```

```python
# most popular product_category state wise
```

<!-- Most popular product category state wisee -->

```python
df_state_category=df.groupby(by=['State','Product_Category']).size().reset_index(name='count')
df_state_category
idx = df_state_category.groupby('State')['count'].idxmax()

# # # Select only the rows with the max count per Zone
df_best_per_state = df_state_category.loc[idx].reset_index(drop=True)

print(df_best_per_state)
```

```python
df_occupation_order_count=df.groupby(by=['Occupation']).size().sort_values(ascending=False).head(10).index
df_top10_occupation=df[df['Occupation'].isin(df_occupation_order_count)]
df_top10_occupation
df_occupation_gender_count = (
    df_top10_occupation
    .groupby(['Occupation', 'Gender'])
    .size()
    .reset_index(name='count')
)
df_occupation_gender_count
plt.figure(figsize=(12, 6))

sns.barplot(
    data=df_occupation_gender_count,
    y='Occupation',
    x='count',
    hue='Gender'
)

plt.tight_layout()
plt.show()
```

```python
# Top3 Occupation and their top5 product categories
df_occupation_amount=df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).head(3).index
df_occupation_amount
df_top3_occupation=df[df['Occupation'].isin(df_occupation_amount)]
df.head()
df_top3_occupation=df_top3_occupation.groupby(by=['Occupation','Product_Category'])['Amount'].sum().reset_index()
df_top3_occupation

df_top3_occpation_top5_category=df_top3_occupation.sort_values(by=['Occupation','Amount'],ascending=[True,False]).groupby('Occupation').head(5)
df_top3_occpation_top5_category
```
